```python
# Dependencies

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import beta
```

```python
a, b = 10,5 
mean, var, skew, kurt = beta.stats(a, b, moments='mvsk')
x = np.linspace(beta.ppf(0.0, 6, 3),beta.ppf(1.00, 1, 10), 100)
```

```python
# Visualizing priors

f, axarr = plt.subplots(3, 3,figsize=(10,8))

axarr[0, 0].plot(x, beta.pdf(x, 23, 7),'b-', lw=4, alpha=0.6)
axarr[0, 0].set_title('sta-->sta')

axarr[0, 1].plot(x, beta.pdf(x, 5, 25),'b-', lw=4, alpha=0.6)
axarr[0, 1].set_title('pre-->sta')

axarr[0, 2].plot(x, beta.pdf(x, 1, 29),'b-', lw=4, alpha=0.6)
axarr[0, 2].set_title('chu-->sta')

axarr[1, 0].plot(x, beta.pdf(x, 2, 28),'b-', lw=4, alpha=0.6)
axarr[1, 0].set_title('sta-->pre')

axarr[1, 1].plot(x, beta.pdf(x, 22, 8),'b-', lw=4, alpha=0.6)
axarr[1, 1].set_title('pre-->pre')

axarr[1, 2].plot(x, beta.pdf(x, 1, 29),'b-', lw=4, alpha=0.6)
axarr[1, 2].set_title('chu-->pre')

axarr[2, 0].plot(x, beta.pdf(x, 5, 25),'b-', lw=4, alpha=0.6)
axarr[2, 0].set_title('sta-->chu')

axarr[2, 1].plot(x, beta.pdf(x, 3, 27),'b-', lw=4, alpha=0.6)
axarr[2, 1].set_title('pre-->chu')

axarr[2, 2].plot(x, beta.pdf(x, 28, 2),'b-', lw=4, alpha=0.6)
axarr[2, 2].set_title('chu-->chu')

# Fine-tune figure; hide x ticks for top plots and y ticks for right plots
plt.setp([a.get_xticklabels() for a in axarr[0, :]], visible=False)
plt.setp([a.get_xticklabels() for a in axarr[1, :]], visible=False)
plt.setp([a.get_yticklabels() for a in axarr[:, 2]], visible=False)
plt.setp([a.get_yticklabels() for a in axarr[:, 1]], visible=False)
plt.show()
```

```python
sta_sta,sta_pre,sta_chu = 68,14,16
pre_sta,pre_pre,pre_chu = 8,87,5
chu_sta,chu_pre,chu_chu = 0,0,0

a, b = 10,5 
mean, var, skew, kurt = beta.stats(a, b, moments='mvsk')
x = np.linspace(beta.ppf(0.0, 6, 3),beta.ppf(1.00, 1, 10), 100)
```

```python
f, axarr = plt.subplots(3, 3,figsize=(10,8))
axarr[0, 0].plot(x, beta.pdf(x, 23+sta_sta, 7+sta_pre+sta_chu),'r-', lw=4, alpha=0.6)
a11 = ((23+sta_sta)/((23+sta_sta)+(7+sta_pre+sta_chu)))
axarr[0, 0].plot(x, beta.pdf(x, 23, 7),'b-', lw=4, alpha=0.6)
axarr[0, 0].set_title('sta-->sta')

axarr[0, 1].plot(x, beta.pdf(x, 5+pre_sta, 25+pre_pre+pre_chu),'r-', lw=4, alpha=0.6)
a12 =  ((5+pre_sta)/((5+pre_sta)+(25+pre_pre+pre_chu)))
axarr[0, 1].plot(x, beta.pdf(x, 5, 25),'b-', lw=4, alpha=0.6)
axarr[0, 1].set_title('pre-->sta')

axarr[0, 2].plot(x, beta.pdf(x, 1+chu_sta, 29+chu_pre+chu_chu),'r-', lw=4, alpha=0.6)
a13 =  ((1+chu_sta)/((1+chu_sta)+(29+chu_pre+chu_chu)))
axarr[0, 2].plot(x, beta.pdf(x, 1, 29),'b-', lw=4, alpha=0.6)
axarr[0, 2].set_title('chu-->sta')

axarr[1, 0].plot(x, beta.pdf(x, 2+sta_pre, 28+sta_chu+sta_sta),'r-', lw=4, alpha=0.6)
a21 = ((2+sta_pre)/((2+sta_pre)+(28+sta_chu+sta_sta)))
axarr[1, 0].plot(x, beta.pdf(x, 2, 28),'b-', lw=4, alpha=0.6)
axarr[1, 0].set_title('sta-->pre')

axarr[1, 1].plot(x, beta.pdf(x, 22+pre_pre, 8+pre_sta+pre_chu),'r-', lw=4, alpha=0.6)
a22 =  ((22+pre_pre)/((22+pre_pre)+(8+pre_sta+pre_chu)))
axarr[1, 1].plot(x, beta.pdf(x, 22, 8),'b-', lw=4, alpha=0.6)
axarr[1, 1].set_title('pre-->pre')

axarr[1, 2].plot(x, beta.pdf(x, 1+chu_pre, 29+chu_sta+chu_chu),'r-', lw=4, alpha=0.6)
a23 =  ((1+chu_pre)/((1+chu_pre)+(29+chu_sta+chu_chu)))
axarr[1, 2].plot(x, beta.pdf(x, 1, 29),'b-', lw=4, alpha=0.6)
axarr[1, 2].set_title('chu-->pre')

axarr[2, 0].plot(x, beta.pdf(x, 5+sta_chu, 25+sta_pre+sta_sta),'r-', lw=4, alpha=0.6)
a31 =  ((5+sta_chu)/((5+sta_chu)+(25+sta_pre+sta_sta)))
axarr[2, 0].plot(x, beta.pdf(x, 5, 25),'b-', lw=4, alpha=0.6)
axarr[2, 0].set_title('sta-->chu')

axarr[2, 1].plot(x, beta.pdf(x, 3+pre_chu, 27+pre_pre+pre_sta),'r-', lw=4, alpha=0.6)
a32 =  ((3+pre_chu)/((3+pre_chu)+(27+pre_pre+pre_sta)))
axarr[2, 1].plot(x, beta.pdf(x, 3, 27),'b-', lw=4, alpha=0.6)
axarr[2, 1].set_title('pre-->chu')

axarr[2, 2].plot(x, beta.pdf(x, 28+chu_chu, 2+chu_pre+chu_sta),'r-', lw=4, alpha=0.6)
a33 =  ((28+chu_chu)/((28+chu_chu)+(2+chu_pre+chu_sta)))
axarr[2, 2].plot(x, beta.pdf(x, 28, 2),'b-', lw=4, alpha=0.6)
axarr[2, 2].set_title('chu-->chu')

# Fine-tune figure; hide x ticks for top plots and y ticks for right plots
plt.setp([a.get_xticklabels() for a in axarr[0, :]], visible=False)
plt.setp([a.get_xticklabels() for a in axarr[1, :]], visible=False)
plt.setp([a.get_yticklabels() for a in axarr[:, 2]], visible=False)
plt.setp([a.get_yticklabels() for a in axarr[:, 1]], visible=False)
plt.show()
```

```python
a_val,b_val,c_val = [],[],[]
# initial state
init_state = np.array([100, 100, 0])
# transition matrix
a = np.array([[ a11,a12 ,a13], [a21, a22,a23], [ a31,a32,a33]])
for x in range(10):
    a_val.append(init_state[0])
    b_val.append(init_state[1])
    c_val.append(init_state[2])
    b = init_state
    init_state = a.dot(b)
# plotting
plt.figure(figsize=(11,8))
plt.plot( [x for x in range(10)], a_val, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4,label='Standard users')
plt.plot( [x for x in range(10)], b_val, marker='o', markerfacecolor='red', markersize=12, color='pink', linewidth=4,label='Premium users')
plt.plot( [x for x in range(10)], c_val, marker='o', markerfacecolor='orange', markersize=12, color='yellow', linewidth=4,label='Churned users')
plt.legend(loc='best')
plt.xlabel('Months')
plt.ylabel('Number of customers')
```
